```python
from flair.data import Corpus
from flair.datasets import CONLL_03
from flair.embeddings import TokenEmbeddings, WordEmbeddings, StackedEmbeddings, FlairEmbeddings, TransformerWordEmbeddings, BytePairEmbeddings
from typing import List

corpus: Corpus = CONLL_03(base_path='/content/drive/MyDrive/res')


tag_type = 'ner'

tag_dictionary = corpus.make_tag_dictionary(tag_type=tag_type)

embedding_types: List[TokenEmbeddings] = [
  WordEmbeddings('glove'),
  FlairEmbeddings('news-forward-fast'),
    FlairEmbeddings('news-backward-fast')
]


embeddings: StackedEmbeddings = StackedEmbeddings(embeddings=embedding_types)

from flair.models import SequenceTagger

tagger: SequenceTagger = SequenceTagger(hidden_size=256,
                                        use_crf=True,
                                        embeddings=embeddings,
                                        tag_dictionary=tag_dictionary,
                                        
                                        tag_type=tag_type)

from flair.trainers import ModelTrainer

trainer: ModelTrainer = ModelTrainer(tagger, corpus)

trainer.train('/content/drive/MyDrive/res/example-ner-flair-glove',
              train_with_dev=True,  
              embeddings_storage_mode='gpu',
              patience=5,
              max_epochs=100)
```
